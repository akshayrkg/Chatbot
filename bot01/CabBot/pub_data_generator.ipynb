```python
import random
import pandas as pd

file=open('pubdata.txt','w+')
status=['Economy','Medium','Luxury']
zone=['North','East','West','South']
cuisine=['SouthIndian','NorthIndian','Vegetarian','Vegan','European','African','Asian']

file.write('zone,status,cuisine,guests\n')
for i in range(1000):
    
    file.write(str(random.choice(zone)+','+random.choice(status)+','+random.choice(cuisine)+','+str(random.randint(1,10))+'\n'))


file.close()
```
